```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_csv("https://raw.githubusercontent.com/mwaskom/seaborn-data/master/diamonds.csv")
data = df[['carat', 'depth', 'price']].copy()
data.rename(columns={'carat': 'House_Area', 'depth': 'No_of_Rooms', 'price': 'House_Price'}, inplace=True)
sns.set(style="whitegrid", palette="coolwarm", context="talk")
pair_plot = sns.pairplot(data, diag_kind='kde', plot_kws={'alpha': 0.7, 's': 60, 'edgecolor': 'k'})
pair_plot.fig.suptitle("Housing Data Analysis: Relationships Between Features", fontsize=16, y=1.02)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
titanic = sns.load_dataset("titanic")
titanic_clean = titanic[['age', 'fare', 'class', 'survived']].dropna()
sns.set(style="whitegrid", palette="coolwarm", context="talk")
pair_plot = sns.pairplot(
    titanic_clean,
    vars=['age', 'fare'],
    hue='survived',
    diag_kind='kde',
    markers=['o', 's'],
    plot_kws={'alpha': 0.7, 'edgecolor': 'k'}
)
pair_plot.fig.suptitle("Titanic Passenger Patterns: Survival vs. Age, Fare, and Class", fontsize=16, y=1.02)
plt.show()
```
