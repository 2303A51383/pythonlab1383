```python
import matplotlib.pyplot as plt
students =["A","B","C","D","E"]
marks=[85, 70 ,90, 60 ,75]
plt.bar(students ,marks,color="orange")
plt.xlabel("students")
plt.ylabel("marks")
plt.title(" Students Perfomance")
plt.show()
```

```python
import matplotlib.pyplot as plt
hours =[2,3,4,5,6,7,8]
scores = [50,55,65,70,75,80,90]
plt.scatter(hours,scores,color="green")
plt.xlabel("hours")
plt.ylabel("Exam scores")
plt.title("Hours vs Exam Scores")
plt.show()
```

```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5]
y = [2,4,6,8,10]
plt.plot(x,y,color="blue")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("Simple Line Plot")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
tips = sns.load_dataset("tips")
sns.boxplot(x="day",y="total_bill",data=tips)
plt.title("Boxplot of Total Bill by Day")
plt.show()
```

```python
import plotly.express as px
dt = px.data.iris()
fig = px.scatter(dt,x="sepal_width",y="sepal_length",color="species",size="petal_length",hover_data=["petal_width"])
fig.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
data = {"month": ["Jan","Feb","Mar","Apr"],
        "Sales": [200,250,300,400]}
df = pd.DataFrame(data)
df.plot(x="month",y="Sales",kind="line",color="blue")
plt.title("Monthly Sales")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.show()
```
